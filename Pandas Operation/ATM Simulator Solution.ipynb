```python
def atm_sim():
    correct_pin = 1234
    balance_amt = 10000
    
    print("**********  Welcome to the ATM Simulator**********  ")
    print(" ")
    entered_pin = int(input("Enter your PIN : "))
    print("\n")
    
    if entered_pin != correct_pin:
        print("\n\nIncorrect PIN !!!\n\nThank You\n\n")
    else:
        print("**********  Welcome to your account  **********")
        print("\n")
        
        while True:
        
            print("Select one from following options : \n\n1)Check Balance\n2)Withdraw money\n3)Deposit money\n4)Exit")

            choose = int(input("\nEnter your choice : "))
            print(" ")
            if choose == 1:
                print("\nYour Balance is : ", balance_amt)
                print(" ")
            elif choose == 2:
                amt_withdrawal = int(input("Enter the amount : \n"))
                if balance_amt < amt_withdrawal:
                    print("\nInsufficient Balance\n")
                    print(" ")
                else:
                    balance_amt = balance_amt - amt_withdrawal
                    print("Balance Amount : ",balance_amt)
                    print(" ")
            elif choose == 3:
                amt_deposit = int(input("Enter the amount : \n"))
                balance_amt = amt_deposit + balance_amt
                print("Your Balance is : ", balance_amt)
                print(" ")
            elif choose == 4:
                print("Thank You")
                
                break
            else:
                print("Invalid Option Selected")
            print("\n")
            
        
        
        
atm_sim()
```
