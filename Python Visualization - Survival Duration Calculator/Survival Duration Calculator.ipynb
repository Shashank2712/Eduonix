```python
class duration_calc():
    def __init__(self, name = '', months = 0, weeks = 0, days=0, hours = 0, mins= 0, secs = 0):
        self.name = name
        self.months = months
        self.weeks = weeks
        self.days = days
        self.hours = hours
        self.mins = mins
        self.secs = secs
        
    def cal_months(self):
        self.months = self.age*12
        print("\nThe no of months for your age are : ", self.months, "months")
        
    def cal_weeks(self):
        self.weeks = self.age*52 #Since there are 52 weeks in one year in general
        print("\nThe no of weeks for your age are : ", self.weeks, "weeks")
        
    def cal_days(self):
        self.days = self.age*52*7 
        print("\nThe no of days for your age are : ", self.days, "days")

    def cal_hours(self):
        self.hours = self.age*52*7*24 
        print("\nThe no of hours for your age are : ", self.hours, "hours")
        
    def cal_mins(self):
        self.mins = self.age*52*7*24*60 
        print("\nThe no of minutes for your age are : ", self.mins,"minutes")
        
    def cal_secs(self):
        self.secs = self.age*52*7*24*60 
        print("\nThe no of seconds for your age are : ", self.secs, "seconds")
        
    def start(self):
        self.name = input("Enter your name : ")
        self.age = int(input("Enter your age: "))
        print("Please choose the time units : \n\n1) Months\n\n2) Weeks\n\n3) Days\n\n4) Hours\n\n5) Minutes\n\n6) Seconds\n\n7) Exit")
        choose = int(input("\nEnter your choice : "))
        if choose == 1:
            self.cal_months()
        elif choose ==2:
            self.cal_weeks()
        elif choose ==3:
            self.cal_days()
        elif choose == 4:
            self.cal_hours()
        elif choose == 5:
            self.cal_mins()
        elif choose == 6:
            self.cal_secs()
        elif choose == 7:
            print("\nThank You")
        else:
            print("\nInvalid Option selected")
        print("\n")
            
            
if __name__ == '__main__':
    duration = duration_calc()
    duration.start()
               
```
